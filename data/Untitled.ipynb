```python
import pandas as pd
h2r = pd.read_json('output.json', orient='columns')
h2r
```

```python
categories = []
for category in h2r['category']:
    for cate in category:
        if cate not in categories:
            categories.append(cate)
categories.sort()
categories
```

```python
content_tags = []
for content in h2r['content']:
    for tag in content:
        if tag not in content_tags:
            content_tags.append(tag)
content_tags.sort()
content_tags
```

```python
# Common categories and content tags
# Surprisingly none
[x for x in content_tags if x in categories]
```

```python
# Init data frames
title_df = pd.DataFrame(h2r["title"])
category_df = pd.DataFrame(h2r["category"])
content_df = pd.DataFrame(h2r["content"])
rating_df = pd.DataFrame(h2r["rating"])
ranking_df = pd.DataFrame(h2r["ranking"], dtype=object)
views_df = pd.DataFrame(h2r["views"])
```

```python
# Sort by rank
ranking = pd.concat([title_df['title'], views_df['views'], ranking_df['ranking'], rating_df['rating'], category_df['category'], content_df['content']], axis=1, keys=['title', 'views', 'ranking', 'rating', 'category', 'content'])
ranking.sort_values(by='views', ascending=False)
```
